```python
import pandas as pd
from sklearn import linear_model

path="C:/Users/Spoorthy M K/Desktop/python1/mutiplelinearregression/Exam/Examm.csv"
data=pd.read_csv(path)
print(data)
```

```python
x=data.drop('FINAL','columns')
y=data.FINAL
print("The x variable contains: ")
print(x)
print("----------------------------------------------------------------")
print("The y variable contains: ")
print(y)
```

```python
model = linear_model.LinearRegression()
model.fit(x,data.FINAL)
```

```python
print("The predicted output is ",model.predict([[73,80,75]]))
print("The intercept value is ",model.intercept_)
print("The coefficient value is ",model.coef_)
print("The accuracy is ",model.score(x,y))
```
