```python
# total number of female athletes in each olympics

female_participants_summer = athletes_df[(athletes_df.Sex == "F") & (athletes_df.Season == "Summer")][["Sex","Year"]]
female_participants_summer = female_participants_summer.groupby("Year").count().reset_index()
female_participants_summer.head()
female_participants_summer.tail()
```

```python
female_participants_winter = athletes_df[(athletes_df.Sex == "M") & (athletes_df.Season == "Winter")][["Sex","Year"]]
female_participants_winter = female_participants_winter.groupby("Year").count().reset_index()
female_participants_winter.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == "F") & (athletes_df.Season == "Summer")]
```

```python
sns.set(style = "darkgrid")
plt.figure(figsize = (20,10))
sns.countplot(x = "Year", data = womenOlympics, palette = "Spectral")
plt.title("Women participation")
```

```python
part = womenOlympics.groupby("Year")["Sex"].value_counts()
plt.figure(figsize = (20, 10))
part.loc[:,"F"].plot()
plt.title("Plot of Female Athletes over time")
```

```python
# Gold medals athletes
gold_medals = athletes_df[(athletes_df.Medal == "Gold")]
gold_medals.head()
```

```python
goldMedals = gold_medals[np.isfinite(gold_medals["Age"])]   #considering the values that are not NaN
goldMedals.head()
```

```python
goldMedals["ID"][goldMedals["Age"] > 60].count()
```

```python
sports = goldMedals["Sport"][goldMedals["Age"]>60]
sports
```

```python
plt.figure(figsize = (10,5))
plt.tight_layout()
sns.countplot(sports)
plt.title("Gold Medals for Athletes over 60 years")
```

```python
#Gold medals from each country
goldMedals.Region.value_counts().reset_index(name = "Medal").head()
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name = "Medal").head(6)
g = sns.catplot(x = "index", y = "Medal", data = totalGoldMedals, height = 5, kind = "bar", palette = "rocket")
g.despine(left = True)
```
